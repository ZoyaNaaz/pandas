```python
data1 = pd.DataFrame({
    'key1': [1,2,3,4,5],
    'key2': [4,5,6,5,7],
    'key3': [3,4,5,7,6]
})
```

```python
data2 = pd.DataFrame({
    'key1': [1,2,36,4,5],
    'key4': [4,5,67,5,7],
    'key5': [37,4,5,7,60]
})
```

```python
data1
```

```python
data2
```

```python
# to apply joins we use merge()
# it apply inner join by default
# fetch value based on common col
pd.merge(data1, data2)
```

```python
# specify how to do a join
pd.merge(data1, data2, how = 'left')
```

```python
pd.merge(data1, data2, how = 'right')
```

```python
pd.merge(data1, data2, how = 'outer')
```

```python
# cross product of data1 values with data2
pd.merge(data1, data2, how = 'cross')
```

```python
# specify col to be merged upon
pd.merge(data1, data2, how='left', left_on='key2', right_on='key4')
```

```python
data3 = pd.DataFrame({
    'key1': [1,2,4,5,6],
    'key2': [4,5,6,7,8],
    'key3': [3,4,5,6,6]
    },
    index=['a', 'b', 'c', 'd', 'e']
)
```

```python
def create_flag(x):
    if x < 10:
        return 'cheap'
    elif x >=10 and x <= 20:
        return 'mid'
    else:
        return 'high'
```

```python
df['Flag_fare'] = df['Fare'].apply(create_flag)
```
